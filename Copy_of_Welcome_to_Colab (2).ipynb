## Initial Data Loading

```
import pandas as pd
df = pd.read_csv('/content/session4_dummy_students_data.csv.xls')
```

## Initial Data Exploration

### Display First Few Rows

```
print('First 5 rows of the DataFrame:')
display(df.head())
```

### Check DataFrame Shape

```
print('\nShape of the DataFrame:')
display(df.shape)
```

### Get DataFrame Information (Data Types, Non-Null Counts)

```
print('\nInformation about the DataFrame (data types, non-null values):')
display(df.info())
```

### View Descriptive Statistics

```
print('\nDescriptive statistics of the DataFrame:')
display(df.describe())
```

## Intermediate NumPy Array Creation

### Math Scores as NumPy Array

```
math_scores_np = df['math'].to_numpy()
print(math_scores_np)
```

### Science Scores as NumPy Array

```
science_scores_np = df['science'].to_numpy()
print(science_scores_np)
```

### English Scores as NumPy Array

```
english_scores_np = df['english'].to_numpy()
print(english_scores_np)
```

### Calculate Average Score (Using NumPy Arrays)

```
avg=(math_scores_np+science_scores_np+english_scores_np)/3
print(avg)
```

### Identify Students with Average Score Greater Than 85

```
students_above_85_avg = df[avg > 85]['name']
print('Students with average score greater than 85:')
print(students_above_85_avg.to_string(index=False))
```

## Adding Calculated Columns to DataFrame

### Add 'average' Column to DataFrame

```
df['average'] = df[['math', 'science', 'english']].mean(axis=1)
df
```

### Add 'result' Column (Pass/Fail) to DataFrame

```
df['result'] = df['average'].apply(lambda x: 'Pass' if x >= 40 else 'Fail')
df
```

## Summary Statistics

### Calculate Class Average

```
class_average = df['average'].mean()
print(f"Class Average: {class_average:.2f}")
```

### Determine Highest and Lowest Averages

```
highest_average = df['average'].max()
lowest_average = df['average'].min()

print(f"Highest Average: {highest_average:.2f}")
print(f"Lowest Average: {lowest_average:.2f}")
```

### Top Student(s) Based on Average Score

```
top_students = df[df['average'] == highest_average]
print("Top Student(s):")
print(top_students)
```

## Specific Student Group Analysis

### Students with Low Attendance (Below 75%)

```
low_attendance = df[df['attendance_percent'] < 75]
print("Students with attendance below 75:")
print(low_attendance)
```

### Students with Fewer Than 6 Assignments Submitted

```
few_assignments = df[df['assignments_submitted'] < 6]
print("Students with fewer than 6 assignments:")
print(few_assignments)
```

## Exporting Results

### Save Processed DataFrame to CSV

```
df.to_csv('session4_student_report.csv', index=False)
print("Report successfully saved!")
```
