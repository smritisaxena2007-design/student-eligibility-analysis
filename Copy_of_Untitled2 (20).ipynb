## Dictionary Operations: Student Data

```python
dic=[
    {"class":"ca","college":"gla","name":"pratibha","city":"mathura","grade":"a+"},
    {"class":"ca","college":"gla","name":"rohan","city": "mathura", "grade": "b+"},
    {'class': 'ca', 'college': 'gla', 'name': 'rahi', 'city': 'mathura', 'grade': 'a+'},
    {'class': 'ca', 'college': 'gla', 'name': 'pari', 'city': 'mathura', 'grade': 'b'},
    {'class': 'ca', 'college': 'gla', 'name': 'palak', 'city': 'mathura', 'grade': 'b'}
]
print(dic)
```

### Iterating through Dictionary

```python
for student in dic:
    if student["grade"] == "a+":
        print(student["name"])
```

```python
for student in dic:
  if student.get("grade") == "a+":
    print(student.get("name"))
```

## Conditional Statements: Even/Odd Check

```python
marks=72
if marks%2==0:
  print("even")
else:
  print("odd")
```

## Functions: Calculating Averages

```python
def calculate_average(marks):
  return sum(marks)/len(marks)
student_marks=[78,84,62]
average=calculate_average(student_marks)
print(average)
```

## Basic Arithmetic: Direct Average Calculation

```python
marks=[78,84,62]
avg=sum(marks)/len(marks)
print(avg)
```
